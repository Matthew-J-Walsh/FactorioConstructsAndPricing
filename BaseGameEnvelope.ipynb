```python
from utils import *
from generators import *
from constructs import *

testing_constuct_A = UncompiledConstruct("test1", {'balls': -69}, {'nuts': 420, 'capybara': -3, 'memepower': -111}, 
                                         {'speed': ['nuts', 'capybara'], 'productivity': ['nuts'], 'consumption': ['memepower']},
                                         [(m,1) for m in data['module'].values()],
                                         base_inputs={'capybara': -9}, cost={'meme': 2}, limit=TechnologicalLimitation([[]]))
testing_constuct_B = UncompiledConstruct("test2", {'ligma': -7}, {'balls': 18, 'crack': -48, 'memepower': -232}, 
                                         {'speed': ['balls', 'crack'], 'productivity': ['balls'], 'consumption': ['memepower']},
                                         [(m,1) for m in data['module'].values()],
                                         base_inputs={'crack': -2}, cost={'doge': 3}, limit=TechnologicalLimitation([[]]))
print(testing_constuct_A)
print(testing_constuct_B)
testing_family_pre, testing_references = generate_all_construct_families([testing_constuct_A, testing_constuct_B])
compiled_construct = Build_New_Construct([c.omni_case for c in testing_family_pre], np.array([3,13]), 
                                         [(testing_references.index('balls'), 1), (testing_references.index('balls'), -1)])
print("============================")
print(testing_references)
print("============================")
print(compiled_construct.ident)
#print([tens for tens in testing_family[0].omni_case.effect_tensors])
phi = np.array([.25,0,0,0,.25,0,0,0,.5, 0,0,.1,0,0,.4,0,.5,0])
theta = np.array([.75,.8589])
print("Theta bounds:")
for half_shell_index in range(compiled_construct.theta_tensors[0].shape[0]):
    value = compiled_construct.theta_tensors[0].to_dense()[half_shell_index]
    value += einstein_summation('ij,j', compiled_construct.theta_tensors[1].to_dense(),
                                theta)[half_shell_index]
    value += einstein_summation('ijk,j,k', compiled_construct.theta_tensors[2].to_dense(),
                                theta, phi)[half_shell_index]
    value += einstein_summation('ijkl,j,k,l', compiled_construct.theta_tensors[3].to_dense(),
                                theta, phi, phi)[half_shell_index]
    value += einstein_summation('ijklm,j,k,l,m', compiled_construct.theta_tensors[4].to_dense(),
                                theta, phi, phi, phi)[half_shell_index]
    print(str(value)+' <= 0 | '+str(eval(str(value)+' <= 0')))
print('Phi bounds:')
for half_shell_index in range(compiled_construct.phi_tensors[0].shape[0]):
    value = compiled_construct.phi_tensors[0].to_dense()[half_shell_index]
    value += einstein_summation('ij,j', compiled_construct.phi_tensors[1].to_dense(),
                                phi)[half_shell_index]
    print(str(value)+' <= 0 | '+str(eval(str(value)+' <= 0')))
print("Cost:")
print(compiled_construct.cost_tensors[0].to_dense() + einstein_summation('ij,j', compiled_construct.cost_tensors[1], phi))
print("Effect evaluation:")
for i in range(len(testing_references)):
    value = compiled_construct.effect_tensors[0].to_dense()[i]
    value += einstein_summation('ij,j', compiled_construct.effect_tensors[1], theta)[i]
    value += einstein_summation('ijk,j,k', compiled_construct.effect_tensors[2], 
                                theta, phi)[i]
    value += einstein_summation('ijkl,j,k,l', compiled_construct.effect_tensors[3], 
                                theta, phi, phi)[i]
    value += einstein_summation('ijklm,j,k,l,m', compiled_construct.effect_tensors[4], 
                                theta, phi, phi, phi)[i]
    if value!=0:
        print(testing_references[i]+': '+str(value))
compiled_construct.freeze(phi)
```

```python
#many test for sparse tensor implementation
from sparsetensors import *

A = np.random.rand(*(100,25,15,5))*(np.random.rand(*(100,25,15,5))>.97)
B = np.random.rand(*(50,20,10,5))*( np.random.rand(*(50,20,10,5))>.95)
C = np.random.rand(*(150,45,25))
Asparse = SparseTensor(A.shape)
Bsparse = SparseTensor(B.shape)
Asparse.from_dense(A)
Bsparse.from_dense(B)
assert np.all(A==Asparse.to_dense())
assert np.all(B==Bsparse.to_dense())
edpnumpy = extra_dimensional_projection_numpy([A,B], [3])
edpmine = extra_dimensional_projection([Asparse,Bsparse], [3])
assert np.all(edpnumpy==edpmine.to_dense())
np.testing.assert_array_almost_equal(einstein_summation("ijkl,ijk", edpnumpy, C),
                                     einstein_summation("ijkl,ijk", edpmine, C))
Ar = SparseTensor(A.shape)
Ar.from_dense(A)
Ar.add_rank(0)
np.sum(Ar.to_dense())
```
